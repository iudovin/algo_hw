1. Решите уравнения в целых числах, используя расширенный алгоритм Евклида:  

	a) $ 238 x + 385 y = 133 $

```python
def gcd(a, b, x, y):
    if a == 0:
        x = 0
        y = 1
        return b, x, y
    x1 = y1 = 0
    d, x1, y1 = gcd(b % a, a, x1, y1)
    x = y1 - (b // a) * x1
    y = x1
    return d, x, y

x = y = 0
solve(238, 385, x, y)
```

2. Вычислите $7^{13} \bmod 167$, используя алгоритм быстрого возведения в степень.

```python
def sq(x):
    return x * x

def pow(a, b):
    if b == 0:
        return 1
    if b % 2 == 0:
        return sq(pow(a, b // 2))
    else:
        return a * sq(pow(a, b // 2))
    
print('Ответ:', pow(7, 13) % 167)
# print(7 ** 13 % 167)
```

3. Доказать корректность рекурсивного алгоритма **Divide** (раздел 1.1., рис. 1.2.) и получить верхнюю оценку на время работы:
        функция Divide(x, y)
        {Вход: n-битовые x и y, причём y 􏰅􏰅⩾ 1.} 
        {Выход: частное и остаток от деления x на y.} 
        если x = 0: вернуть (q, r) = (0, 0)
        (q, r) ← Divide(⌊x / 2⌋, y)
        q ← 2 · q, r ← 2 · r
        если x нечётно: r ← r + 1
        если r ⩾ 􏰅y: r ← r − y,q ← q + 1
        вернуть (q, r)
$\blacktriangle\hspace{10pt}$ Данный алгоритм выполняет деление "уголком" в двоичной системе счисления. Сначала нужно сократить двоичную запись $a$ вычеркиванием цифр с конца до минимальной длины, при которой оно делится на $b$. Операция деления даст временные частное $q = 1$ и остаток $r$ (алгоритм сперва дойдет до первой цифры, после чего "соберет" остаток, дописывая цифры справа). Затем нужно приписать к полученному остатку следующую цифру числа $a$: если $a$ с ней в конце было нечетным, то она равна $1$, иначе $0$. Перед этим нужно добавить разряд к частному $q$ и остатку $r$, умножив их на $2$, тогда можно будет просто прибавить к $r$ единицу, если $a$ было нечетным. Начиная с этого момента, пока алгоритм не дошел до конца числа $a$, остаток $r$ с добавленным разрядом будет делиться на $y$, значит к $q$ нужно прибавить $1$ (до этого там был приписан $0$ умножением на $2$), а в качестве остатка $r$ записать разность исходного остатка $r$ и $y$. Когда алгоритм дойдет до конца числа. $a$, это условие выполняться уже не будет и в $r$ сохранится искомый остаток, а в $q$ $-$ частное. $\hspace{10pt}\blacksquare$  
Асимптотика. Алгоритм рекуррентно "спускается" до наибольшего разряда $a$, а затем "поднимается" обратно, делая на каждом уровне фиксированное число действий. Количество операций можно оценить как $C\cdot\log_2n$, где $n$ $-$ длина двоичного представления $a$. Таким образом, ответ: $O(\log n)$.

```python
def divide(x, y):
    if x == 0:
        return 0, 0
    q, r = divide(x // 2, y)
    q *= 2
    r *= 2
    if x % 2 != 0:
        r += 1
    if r >= y:
        r -= y 
        q += 1
    return q, r

divide(22, 7)
```
