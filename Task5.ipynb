1. Дан массив длины $n$, состоящий только из нулей и единиц. Предложите линейный алгоритм сортировки данного массива. 

    1) Нам достаточно посчитать количество нулей и единиц. Для этого можно создать два счетчика: один сохраняет число нулей, второй $-$ число единиц. На самом деле будет достаточно одного счетчика, т. к. если мы знаем длину массива, то можем определить значение второго счетчика. Посчитаем число нулей в массиве, а затем все элементы на позициях до этого числа заменим на нули, а остальные $-$ на единицы. Нужно пройти все элементы от начала до конца, поэтому асимптотика $-$ $O(n)$.
    
    2) Есть более быстрый алгоритм, изменяющий массив "на месте" и работающий *не более чем* за $n$ операций. Нужно запомнить позицию последнего нуля и первой единицы в массиве (в случае сортировки по возрастанию, иначе наоборот). Пока первая единица стоит в массиве левее последнего нуля, он не отсортирован. Алгоритм находит последний ноль и первую единицу и меняет их местами, затем повторяет это действие, пока массив не будет отсортирован. Когда последний ноль станет левее первой единицы, они будут стоять друг за другом, а значит, алгоритм остановится. При этом индескы первой единицы и последнего нуля пройдут в сумме $n$ значений, т. к. они начинают с концов массива и заканчивают, когда встретятся. Асимптотика: $O(n)$.

```python
# создание случайных массивов

from random import randint as ri
n = 100
a = [ri(0, 1) for _ in range(n)]
b = a.copy()

print(*a)
```

```python
# обычный способ

%time
n0 = 0
for i in range(n):
    if a[i] == 0:
        n0 += 1
for i in range(n):
    if i < n0:
        a[i] = 0
    else:
        a[i] = 1

print(*a)
```

```python
# быстрый способ in-place

%time
def swap(a, i, j):
    if i != j:
        a[i], a[j]  = a[j], a[i] 

f1 = 0
l0 = n - 1

while True:
    while b[f1] != 1:
        f1 += 1
    while b[l0] != 0:
        l0 -= 1
    if f1 < l0:
        swap(b, f1, l0)
    else:
        break

print(*b)
```
